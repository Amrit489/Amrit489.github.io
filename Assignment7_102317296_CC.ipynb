```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# Set random seed (replace with your roll number)
np.random.seed(12345)

# Generate synthetic sales data for 12 months
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales_data_array = np.random.randint(1000, 5000, size=(12, 4))

# Convert to DataFrame
sales_df = pd.DataFrame(sales_data_array, columns=categories, index=months)

# Display first 5 rows and summary statistics
print(sales_df.head())
print(sales_df.describe())

# Calculate total sales per category and per month
total_sales_per_category = sales_df.sum()
total_sales_per_month = sales_df.sum(axis=1)

# Calculate average sales growth between consecutive months
sales_growth = sales_df.pct_change().mean()

# Add Total Sales and Growth Rate columns
sales_df["Total Sales"] = total_sales_per_month
sales_df["Growth Rate"] = sales_df["Total Sales"].pct_change() * 100

# Apply discount based on roll number (even = Electronics 10%, odd = Clothing 15%)
roll_number = 12345  # Replace with actual roll number
if roll_number % 2 == 0:
    sales_df["Electronics"] *= 0.9
else:
    sales_df["Clothing"] *= 0.85

# Visualization
plt.figure(figsize=(10, 5))
sales_df[categories].plot(marker='o')
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend(title="Category")
plt.grid()
plt.show()

# Box plots for sales distribution
plt.figure(figsize=(10, 5))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales Units")
plt.show()

# Display summary results
print("Total Sales Per Category:")
print(total_sales_per_category)
print("\nTotal Sales Per Month:")
print(total_sales_per_month)
print("\nAverage Sales Growth Per Category:")
print(sales_growth)
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row.
iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])


abs_array = np.abs(array)


flat_percentiles = np.percentile(array, [25, 50, 75])
col_percentiles = np.percentile(array, [25, 50, 75], axis=0)
row_percentiles = np.percentile(array, [25, 50, 75], axis=1)


flat_mean = np.mean(array)
col_mean = np.mean(array, axis=0)
row_mean = np.mean(array, axis=1)

flat_median = np.median(array)
col_median = np.median(array, axis=0)
row_median = np.median(array, axis=1)

flat_std = np.std(array)
col_std = np.std(array, axis=0)
row_std = np.std(array, axis=1)

print("Element-wise Absolute Value:\n", abs_array)
print("\nPercentiles (25th, 50th, 75th):")
print("Flattened:", flat_percentiles)
print("Columns:\n", col_percentiles)
print("Rows:\n", row_percentiles)
print("\nMean:")
print("Flattened:", flat_mean)
print("Columns:", col_mean)
print("Rows:", row_mean)
print("\nMedian:")
print("Flattened:", flat_median)
print("Columns:", col_median)
print("Rows:", row_median)
print("\nStandard Deviation:")
print("Flattened:", flat_std)
print("Columns:", col_std)
print("Rows:", row_std)
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceil_values = np.ceil(a)
trunc_values = np.trunc(a)
round_values = np.round(a)

print("Floor Values:", floor_values)
print("Ceiling Values:", ceil_values)
print("Truncated Values:", trunc_values)
print("Rounded Values:", round_values)
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
def swap_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp

lst = [1, 2, 3, 4, 5]
swap_elements(lst, 1, 3)
print(lst)
```

Q.5 Write a Python program to swap two elements in a set by converting it to a list first.

```python
def swap_elements_in_set(s, elem1, elem2):
    lst = list(s)
    if elem1 in lst and elem2 in lst:
        i, j = lst.index(elem1), lst.index(elem2)
        lst[i], lst[j] = lst[j], lst[i]
    return set(lst)

s = {1, 2, 3, 4, 5}
s = swap_elements_in_set(s, 2, 4)
print(s)
```
