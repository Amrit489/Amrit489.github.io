Q.6 Create a sample dataset (employees.csv) containing information about employees in a company.

WAP to determine the following using dataset in Q6:
a) Shape(numberofrowsandcolumns)oftheDataFrame.
b) SummaryoftheDataFramethatincludesthedatatypesandnon-nullcountsfor
each column.
c) Generatedescriptivestatistics.
d) Display the first 5 rows and last 3 rows of the dataset.
e) Calculatethefollowingstatisticsfromthedataset:
    i. The average salary of employees.
    ii. The total bonus paid to all employees.
    iii. The youngest employee's age.
    iv. The highest performance rating.
f) Sort the DataFrame by the Salary column in descending order.
g) Add a new column that categorizes employees based on their performance rating:
    i. Excellent for ratings >= 4.5
    ii. Good for ratings >= 4.0 but < 4.5
    iii. Average for ratings < 4.0
h) IdentifymissingvaluesintheDataFrame.
i) Rename the Employee_ID column to ID.
j) Find all employees who:
    i. Have more than 5 years of experience.
    ii. Belong to the IT department.
k) Modifythedatasetbyaddinganewcolumn,Tax,whichdeducts10%ofthe Salary.
l) Save the modified DataFrame (with added columns) to a new CSV file.

```python
import pandas as pd
table={'Employee_ID':[101,102,103,104,105],
       'Name':['Amrit','Kritika','Aadar','Diptangshu','Aman'],
       'Department':['Senior Designer','Manager','IT','IT','Marketing'],
       'Age':[21,20,19,21,22],
       'Salary':[450000,500000,215000,215000,450000],
       'YearOfExp.':[4,4,2,2,4],
       'DOJ':['2023-11-13','2024-07-22','2024-12-09','2024-12-03','2023-12-13'],
       'Gender':['Male','Female','Male','Male','Male'],
       'Bonus':[80000,30000,15000,20000,50000],
       'Rating':[4.9,5,4,4,4.9]}
result=pd.DataFrame(table)
print(result)
print('\n')

result.to_csv('employees.csv',index=False)

print('Shape:',result.shape)

print(result.info())
print('\n')

print(result.describe())
print('\n')

print(result.iloc[0:4])
print('\n')

print(result.iloc[2:4])
print('\n')

print('Average Salary:',result['Salary'].mean())
print('\n')

print('Total Bonus:',result['Bonus'].sum())
print('\n')

print('Youngest Employee:',result['Age'].min())
print('\n')

print('Highest Rating:',result['Rating'].max())
print('\n')

result_sort=result.sort_values(by='Salary',ascending=False)
print(result_sort)
print('\n')

result["Performance"] = pd.cut(
    result["Rating"],
    bins=[0, 4.0, 4.5, 5],
    labels=["Average", "Good", "Excellent"]
)
print(result[["Name", "Performance"]])
print('\n')

result['Performance'] = result['Rating'].apply(
    lambda x: 'Excellent' if x >= 4.5 else 'Good' if x >= 4.0 else 'Average'
)

print(result)
print('\n')

print(result.isnull().sum())
print('\n')

result.rename(columns={'Employee_ID':'ID'}, inplace=True)
print(result)
print('\n')

sort_emp=result[(result['YearOfExp.']>5)&(result['Department']=='IT')]
print(sort_emp)
print('\n')

result['Tax']=result['Salary']*0.1
print(result)

result.to_csv('modified_employees.csv',index=False)
```
